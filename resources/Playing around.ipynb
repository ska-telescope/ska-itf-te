## Test plot with example

```python
# importing the required module
import matplotlib.pyplot as plt
  
# x axis values
x = [1,2,3]
# corresponding y axis values
y = [2,6,1]
  
# plotting the points 
plt.plot(x, y)
  
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
  
# giving a title to my graph
plt.title('My first graph!')
  
# function to show the plot
plt.show()
```

## Import csv using dictionary and plot

```python
import csv

# open the file in read mode
filename = open('Temperature graph import.csv', 'r')

# creating dictreader object
file = csv.DictReader(filename)

# create empty x and y lists    
xval_list = []
yval_list = []

# Step through each time, temp entry in the csv file and add to x and y lists
for col in file:
    # print(f'\t At {col["Time"]} hours, Temperature was {col["Temperature (degC)"]} deg C.')
    x = int(col['Time'])
    y = float(col['Temperature (degC)'])
    xval_list.append(x)
    yval_list.append(y)
    
# print(xval_list)
# print(yval_list)  

# Plot
import matplotlib.pyplot as plt
plt.plot(xval_list, yval_list)
plt.xlabel('Time (hour)')
plt.ylabel('Temperature (degC)')
plt.title('Temperature at Time of Day')
plt.show()
```
